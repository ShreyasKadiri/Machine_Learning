```python
import easyocr
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread("/content/Invoice.png", cv2.IMREAD_COLOR)
plt.rcParams["figure.figsize"] = (30,30)
plt.imshow(image)
plt.show()
```

```python
reader = easyocr.Reader(['en']) 
result = reader.readtext('/content/Invoice.png')
```

```python
for item in result:
  print(item)
```
