```python
#Word Embedding Techniques using Embedding Layer in Keras
```

```python
#Importing the libraries
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sentences = [  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
### Vocabulary size
vocabulary_size=10000
```

```python
#One Hot Representation
```

```python
onehot_representation = [one_hot(words,vocabulary_size)for words in sentences] 
print(onehot_representation)
```

```python
#Word Embedding Representation

import numpy as np
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
sentence_length=8
embedded_sentences=pad_sequences(onehot_representation,padding='pre',maxlen=sentence_length)
print(embedded_sentences)
```

```python
dimensions = 10
model=Sequential()
model.add(Embedding(vocabulary_size,10,input_length=sentence_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_sentences))
```
